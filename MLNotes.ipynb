**Visualization with Seaborn**

```python
#Visualization with Seaborn
import pandas as pd
df = pd.read_csv("nba.csv")
#Line Plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x = df['Age'], y = df['Weight'], hue = df['Position'])
plt.show()
#Hue is meant for categorization
```

```python
#Scatter Plot
sns.scatterplot(x = df['Age'], y = df['Weight'], hue = df['Position'])
plt.show()
```

```python
#Box Plot
sns.boxplot(df['Age'])
```

```python
#Histogram
sns.histplot(df['Position'], kde = True)
plt.title('Distribution of total bill')
plt.show()
```

```python
#Create a pair plot
import seaborn as sns
sns.pairplot(df, hue = 'position')
plt.title("Pair plot of tips data")
plt.show()
```
